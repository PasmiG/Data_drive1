* Brand volkswagen is sold the most on ebay, follwed by bmw and opel.

```python
brands_top_twenty = autos["brand"].value_counts().head(20)
mean_val = {}
brands = brands_top_twenty.index

for br in brands:
    avg = autos.loc[autos["brand"] == br, "price"].mean()
    mean_val[br] = round(avg, 2)
print(mean_val)    

```

* Even though volkswagen is the top sold cars in ebay, citroen has the highest selling price.

```python
brands_top_twenty = autos["brand"].value_counts().head(20)
mean_val = {}
mean_mil = {}
brands = brands_top_twenty.index

for br in brands:
    avg = autos.loc[autos["brand"] == br, "price"].mean()
    avg_mil = autos.loc[autos["brand"]== br, "odometer_km" ].mean()
    mean_val[br] = round(avg, 2)
    mean_mil[br] = avg_mil
```

```python
mean_val_series = df.Series(mean_val) 
mean_mil_series = df.Series(mean_mil)

df_mean = df.DataFrame(mean_val_series,columns=['mean_values'])
df_mean
```
