```python
def cat_level(c):
    if c == "1-2":
        return 1
    elif c == "3-4":
        return 3
    elif c == "5-6":
        return 5
    elif c == "7-10":
        return 7
    elif c == "11-20":
        return 12
    elif c== "Less than 1 year":
        return 1
    elif c== "More than 20 years":
        return 21
    else:
        return c
    
```

```python
combined_updated1['institute_service1']=combined_updated1['institute_service'].apply(cat_level)
```

```python
combined_updated1['institute_service1'].value_counts()
```

```python
def update_ser (val):
    if pd.isnull(val):
        return np.nan
    elif val < 3.0:
        return 'New'
    elif (val >= 3.0) & (val <= 6.0):
        return 'Experienced'
    elif (val >= 7.0) & (val <=10.0):
        return 'Established'
    else:
        return 'Veteran'
```
